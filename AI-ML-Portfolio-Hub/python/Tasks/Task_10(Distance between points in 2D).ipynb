```python
# Import the math module to use mathematical functions like sqrt.
import math
```

```python
class Point:
    def __init__(self, x, y):
        """
        Describtion of this function:
        #Initialize a Point with x and y coordinates.
        # help me to take instance of Point class
        #Argument:
         1)first argument : x
                1)type:int
         2)second argument : y
                1)type:int
        # Return : 
         1)Return nothing
        """
#########################################################################################
        # Validate that the provided coordinates are within the specified range
        if not self.validate_coordinates(x, y):
            raise ValueError("Coordinates must be in the range -50 <= x, y <= 50.")
#########################################################################################

        self.x = x  # Assign the x coordinate
        self.y = y  # Assign the y coordinate

    @staticmethod
    def validate_coordinates(x, y):
        """
        Describtion of this function:
        #Check if coordinates are within the valid range of -50 to 50.
        #Argument:
         1)first argument : x
                1)type:int
         2)second argument : y
                1)type:int
        # Return : 
         1)Return True if valid, else False
        """
        return -50 <= x <= 50 and -50 <= y <= 50  # Return True if valid, else False

    def distance_to(self, other):
        """
        Describtion of this function:
        #Calculate the distance to another Point.
        #Argument:
         1)first argument : other
                1)type:Point
        # Return : 
         1)Return the calculated distance
        """
        # Check if the other object is an instance of Point
        if not isinstance(other, Point):
            raise TypeError("Distance can only be calculated to another Point.")
        # Calculate the Euclidean distance using the distance formula
        distance = math.sqrt((other.x - self.x) ** 2 + (other.y - self.y) ** 2)
        return distance  
```

```python
def compare_results(expected, actual, tolerance=1e-6):
    """
    Describtion of this function:
    #Compare expected and actual results within a specified tolerance.
    #Argument:
         1)first argument : expected
                1)type:int
         2)second argument : actual
                1)type:int
         3) third argument : tolerance
                1)default:1e-6

    # Return : 
         1)Return True if within tolerance, else False
    """
    return abs(expected - actual) <= tolerance  
```

```python
def main():
  
    while True: 
        try:
            # Input for Point A
            x_a = float(input("Enter x coordinate for Point A (-50 to 50) or type 'exit' to quit: "))
            print(f"you are enter the number of first point (A) of X: {x_a} ")
            print()
            if x_a == 'exit':
                break  # Exit the loop if the user types 'exit'
            y_a = float(input("Enter y coordinate for Point A (-50 to 50): "))
            print(f"you are enter the number of first point (A) of Y: {y_a} ")
            print()
            point_a = Point(x_a, y_a)  # Create Point A

            # Input for Point B
            x_b = float(input("Enter x coordinate for Point B (-50 to 50): "))
            print(f"you are enter the number of Second point (B) of X: {x_b} ")
            print()
            y_b = float(input("Enter y coordinate for Point B (-50 to 50): "))
            print(f"you are enter the number of Second point (B) of Y: {y_b} ")
            print()
            point_b = Point(x_b, y_b)  # Create Point B
            

            # Calculate distance between Point A and Point B
            distance = point_a.distance_to(point_b)
            print(f"The distance between Point A and Point B is: {distance:.6f}")
            print()

            # Example for comparing results
            expected_distance = float(input("Enter the expected distance for validation: "))
            print(f"you are enter the expected_distance: {expected_distance}")
            print()
            # Compare calculated distance with expected distance
            if compare_results(expected_distance, distance):
                print("The calculated distance is within the acceptable range of the expected distance.")
            else:
                print("The calculated distance is outside the acceptable range of the expected distance.")
            print()

        except ValueError as ve:
            print(f"Value Error: {ve}")  # Handle invalid input for coordinates
        except TypeError as te:
            print(f"Type Error: {te}")  # Handle type errors, such as invalid distance calculation
        except Exception as e:
            print(f"An error occurred: {e}")  # Handle any other unexpected errors
        finally:
            # Ask the user if they want to continue or exit
            continue_prompt = input("Do you want to calculate another distance? (yes/no): ").strip().lower()
            if continue_prompt != 'yes':
                print("Exiting the program. Goodbye!")  # Exit message
                break  # Exit the while loop
            else :
                print("you are choice 'Yes',so I will continue the program")
            print()
```

```python
# Entry point for the program
if __name__ == "__main__":
    main()  # Call the main function to start the program
```
