```python
def replace_with_vowels(arr):
    """
    Replaces ASCII values in the input array with corresponding lowercase vowels.
    
    Parameters:
    arr (list of int): A list of integers representing ASCII values.

    Returns:
    list: The modified list where ASCII values for vowels are replaced with the vowels themselves.
    
    """

    # Define a mapping of ASCII values to lowercase vowels
    # I find The number of ASCII character from Google 
    # link:
    #https://www.sciencebuddies.org/science-fair-projects/references/ascii-table

    
    vowels = {
        97: 'a',  # ASCII for 'a'
        101: 'e', # ASCII for 'e'
        105: 'i', # ASCII for 'i'
        111: 'o', # ASCII for 'o'
        117: 'u'  # ASCII for 'u'
    }
    
    # Initialize an empty list to store the ASCII values of found vowels
    vowel_numbers = []
    
    # Iterate through the array
    for i in range(len(arr)):
        # Check if the current number is in the vowels mapping
        if arr[i] in vowels: 
            # Replace the number with its corresponding vowel
            arr[i] = vowels[arr[i]]
            
            # Append the ASCII value of the found vowel to the list
            vowel_numbers.append(ord(arr[i]))  # Use ord() to get the ASCII value of the vowel
            
    # Print the ASCII values of the vowels found
    print(f"The ASCII values of the vowels found are: {vowel_numbers}")
    
    # Return the modified array
    return arr
```

```python
def main():
    """
    Main function to interact with the user, allowing them to input numbers,
    which will be processed to replace ASCII values with vowels.
    """
    while True:
        # Display a menu for user options
        print("\nMenu:")
        print("1. Enter numbers to replace with vowels")
        print("2. Exit")
        
        # Get user's choice
        choice = input("Choose an option (1 or 2): ").strip()
        print(f"you are enter Choice : {choice}")
        
        if choice == '1':
            # Prompt user to enter ASCII values
            user_in = input("Enter numbers separated by spaces (ASCII values): ")
            print("\nBefore Strip:")
            print(f"you are enter numbers separated by spaces: {user_in}")
            print()

            user_input =user_in .strip()
            print("\nAfter Strip:")
            print(f"you are enter numbers separated by spaces:{user_input}")
            print()



            
            # Check if input is empty
            if not user_input:
                print("Input cannot be empty. Please enter numbers.")
                continue
            
            # Split the input into parts for validation
            input_parts = user_input.split()
            print("\nAfter Split:")
            print(f"thr number after Split:{input_parts}")
            print()
            valid_input = True
            
            # Validate each part to ensure it is a digit
            for part in input_parts:
                if not part.isdigit():  # Check if each part is a digit
                    print(f"Invalid input: '{part}' is not a valid number.")
                    valid_input = False
                    break
            
            # If input is valid, process it
            if valid_input:
                # Convert the valid input parts to a list of integers
                numbers = list(map(int, input_parts))
                
                # Call the function to replace numbers with corresponding vowels
                result = replace_with_vowels(numbers)
                
                # Print the resulting array
                print("Resulting array:", result)
        
        elif choice == '2':
            # Exit the program with a goodbye message
            print("Exiting the program. Goodbye!")
            break
        
        else:
            # Handle invalid menu option
            print("Invalid choice. Please select option 1 or 2.")
```

```python
# Run the main function if this script is executed
if __name__ == "__main__":
    main()
```
