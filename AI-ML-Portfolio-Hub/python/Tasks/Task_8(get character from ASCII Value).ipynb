```python
def number_to_ascii(num):
    """Convert a number to its corresponding ASCII character."""
    # Check if the input is either an integer or a float
    if not isinstance(num, (int, float)):
        raise TypeError("Input must be an integer or a float representing an integer.")
    
    # Convert float to int if it is an integer value
    if isinstance(num, float) and num.is_integer():
        num = int(num)
    
    # Check if the number is within the valid ASCII range (0-127)
    if 0 <= num <= 127:
        return chr(num)  # Convert number to ASCII character
    else:
        raise ValueError("Input must be an integer between 0 and 127.")
```

```python
def main_of_ascii(input_num):
    """Handle the conversion of a number to an ASCII character and print the result."""
    try:
        # Attempt to convert the input number to an ASCII character
        result = number_to_ascii(input_num)
        print(f"The ASCII character for {input_num} is: {result}")
    except (ValueError, TypeError) as e:
        print(e)  # Print any errors encountered during conversion
```

```python
def input_num_to_int(input_num):
    """Convert the input to an integer and handle invalid inputs."""
    try:
        input_num = int(input_num)  # Try converting to integer
        return input_num
    except ValueError:
        print("Invalid input. Please enter a valid integer.")  # Handle conversion error
        return None  # Return None if conversion fails
```

```python
def main():
    """Main function to run the ASCII conversion program."""
    while True:
        logout = False  # Control variable for logging out
        print(".... choice from list ....")
        print("1. Convert a number to its corresponding ASCII character.")
        print("2. Exit")
        print()
        
        choice = input("Enter your choice (1/2): ")  # Get user choice
        print(f"You entered the choice: {choice}")
        print()

        if choice == '1':  # User wants to convert a number
            while not logout:  # Continue until logged out
                print(".... choice from list ....")
                print("1. Numeric character (0-9)")
                print("2. English character (A-Z)")
                print("3. Special character (!@#$)")
                print("4. Logout ")
                print()

                choice_1 = input("Enter your choice (1/2/3/4): ")  # Get character type choice
                print(f"You entered the choice : {choice_1}")
                print()
                
                if choice_1 == '1':  # Numeric characters
                    print("You must enter the number from '48' to '57'")
                    input_num = input("Please enter Number: ")
                    print(f"You are entered the number : {input_num}")
                    input_num = input_num_to_int(input_num)  # Convert input to int
                    
                    # Check if input is within valid range
                    if input_num is not None and 48 <= input_num <= 57:
                        main_of_ascii(input_num)  # Convert to ASCII
                    else:
                        print("Please enter a number from '48' to '57' to print numeric character (0-9)")
                        print("Please try again")
                    print()

                elif choice_1 == '2':  # English characters
                    while not logout:  # Nested loop for upper/lower case selection
                        print(".... choice from list ....")
                        print("1. Convert number to upper character (A-Z)")
                        print("2. Convert number to lower character (a-z)")
                        print("3. Logout ")
                        choice_2 = input("Enter your choice (1/2/3): ")  # Get sub-choice
                        print(f"You entered the choice: {choice_2}")
                        print()

                        if choice_2 in ['1', '2']:  # Handle upper/lower case
                            input_num = input("Please enter Number: ")
                            print(f"You are entered the number : {input_num}")
                            input_num = input_num_to_int(input_num)  # Convert to int

                            if choice_2 == '1':  # Uppercase characters
                                print("You must enter the number from '65' to '90'")
                                if input_num is not None and 65 <= input_num <= 90:
                                    main_of_ascii(input_num)  # Convert to ASCII
                                else:
                                    print("Please enter a number from '65' to '90' to print upper character (A-Z)")
                                    print("Please try again")
                                print()
                            elif choice_2 == '2':  # Lowercase characters
                                print("You must enter the number from '97' to '122'")
                                if input_num is not None and 97 <= input_num <= 122:
                                    main_of_ascii(input_num)  # Convert to ASCII
                                else:
                                    print("Please enter a number from '97' to '122' to print lower character (a-z)")
                                    print("Please try again")
                                print()
                        elif choice_2 == '3':  # Logout option
                            print("You are logged out")
                            logout = True   # Set logout flag to True
                            break  
                        else:
                            print("Invalid choice. Please enter 1, 2, or 3.")
                            continue

                elif choice_1 == '3':  # Special characters
                    print("You must enter the number from '33' to '47'")
                    print("You may enter the number from '123' to '127'")
                    input_num = input("Please enter Number: ")
                    input_num = input_num_to_int(input_num)  # Convert input to int
                    if input_num is not None:
                        main_of_ascii(input_num)  # Convert to ASCII
                    else:
                        print("Please try again.")
                    print()

                elif choice_1 == '4':  # Logout option
                    print("You are logged out")
                    logout = True  
                    break
                else:
                    print("Invalid choice. Please try again.")
                print()

        elif choice == '2':  # Exit option
            print("Exiting the program.")
            break  # Exit the main loop
        else:
            print("Invalid choice. Please enter 1 or 2.")
```

```python
#Run the program
if __name__ == "__main__":
    main()  
```
